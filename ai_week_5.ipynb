```python
#-----------------------------------------------------------
# Control DC motor
#-----------------------------------------------------------
import skfuzzy as fuzz
import numpy as np
from skfuzzy import control as ctrl

setpoint = ctrl.Antecedent(np.arange(12,32,1),'setpoint')
sensor = ctrl.Antecedent(np.arange(-3,60,1),'sensor')
pwm = ctrl.Consequent(np.arange(0,255,1),'pwm')
# độ phân giải sensor
setpoint['short'] = fuzz.trimf(setpoint.universe,[12,12,22])
setpoint['medium'] = fuzz.trimf(setpoint.universe,[12,22,32])
setpoint['high'] = fuzz.trimf(setpoint.universe,[22,32,32])
# độ phân giải sensor
sensor['very short'] = fuzz.trimf(sensor.universe,[-3,-3,31])
sensor['short'] = fuzz.trimf(sensor.universe,[-3,15,31])
sensor['medium'] = fuzz.trimf(sensor.universe,[15,31,47])
sensor['high'] = fuzz.trimf(sensor.universe,[31,47,60])
sensor['very high'] = fuzz.trimf(sensor.universe,[47,60,60])
# độ phân giải pwm
pwm['very low'] = fuzz.trimf(pwm.universe,[0,0,64])
pwm['low'] =fuzz.trimf(pwm.universe,[0,64,127])
pwm['medium'] = fuzz.trimf(pwm.universe,[0,127,255])
pwm['high'] = fuzz.trimf(pwm.universe,[127,191,255])
pwm['very high'] = fuzz.trimf(pwm.universe,[191,191,255])
# luật
rule1=ctrl.Rule(setpoint['short']&sensor['very short'],pwm['very low'])
rule2=ctrl.Rule(setpoint['short']&sensor['short'],pwm['low'])
rule3=ctrl.Rule(setpoint['short']&sensor['medium'],pwm['medium'])
rule4=ctrl.Rule(setpoint['short']&sensor['high'],pwm['medium'])
rule5=ctrl.Rule(setpoint['short']&sensor['very high'],pwm['high'])

rule6=ctrl.Rule(setpoint['medium']&sensor['very short'],pwm['low'])
rule7=ctrl.Rule(setpoint['medium']&sensor['short'],pwm['low'])
rule10=ctrl.Rule(setpoint['medium']&sensor['medium'],pwm['medium'])
rule8=ctrl.Rule(setpoint['medium']&sensor['high'],pwm['high'])
rule9=ctrl.Rule(setpoint['medium']&sensor['very high'],pwm['high'])

rule11=ctrl.Rule(setpoint['high']&sensor['very short'],pwm['low'])
rule12=ctrl.Rule(setpoint['high']&sensor['short'],pwm['medium'])
rule13=ctrl.Rule(setpoint['high']&sensor['medium'],pwm['high'])
rule14=ctrl.Rule(setpoint['high']&sensor['high'],pwm['high'])
rule15=ctrl.Rule(setpoint['high']&sensor['very high'],pwm['very high'])

pwmctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15])
pwming= ctrl.ControlSystemSimulation(pwmctrl)
```

```python
pwming.input['setpoint'] = 20
pwming.input['sensor'] = 40
pwming.compute()
print(pwming.output['pwm'])
pwm.view(sim = pwming)
setpoint.view()
sensor.view()
pwm.view()
```
