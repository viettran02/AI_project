```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

url ='1.01. Simple linear regression.csv'
data = pd.read_csv(url)
data.head(21)
data.hist()
plt.show()

# Hiển thị dữ liệu trên biểu đồ scatter
plt.scatter(data['SAT'], data['GPA'])
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)

# Tính toán và vẽ đường hồi quy trên biểu đồ
x = data['SAT']
y = data['GPA']
coefficients = np.polyfit(x, y, 1)
m = coefficients[0]
b = coefficients[1]
plt.plot(x, m*x + b, color='r')

# Hiển thị biểu đồ với đường hồi quy
plt.show()

# Tính toán hệ số tương quan
correlation = data.corr()
print(correlation)

```

```python
from sklearn import linear_model
import numpy as np

# Chuyển đổi dữ liệu thành mảng numpy và định dạng lại kích thước
x = np.array(data['SAT']).reshape(-1, 1)
y = np.array(data['GPA']).reshape(-1, 1)

# Tạo một đối tượng mô hình hồi quy tuyến tính và sử dụng dữ liệu để huấn luyện mô hình
regression_model = linear_model.LinearRegression()
regression_model.fit(x, y)

print(regression_model.intercept_)
print(regression_model.coef_)

#Y = 0.27504038*x + 0.00165569
```

```python
import pandas as pd
url ='mtcars.csv'
data = pd.read_csv(url)

data.head(21)
data.hist()
plt.show()


correlation = data.corr()
print(correlation)
```

```python

X = data.drop(data.columns[[0,2]], axis = 1)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black')

correlation = data.corr()
print(correlation)
```

```python
import matplotlib.pyplot as plt
from sklearn import linear_model

model = linear_model.LinearRegression()
X = pd.DataFrame(data['wt'])
y = data['mpg']
model.fit(X, y)

#plt.scatter(X, y)
#plt.plot(X, model.predict(X), color='red')
#plt.xlabel('Weight')
#plt.ylabel('MPG')
#plt.show()

print(model.intercept_)
print(model.coef_)
# Độ chính xác
model.score(X,y)
# Sai số
y_pred = model.predict(X)
residual = data['mpg'] - y_pred
residual.describe()
```

```python
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black')
plt.plot(data['wt'], y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'],data['wt']**2]).T
y = data['mpg']
model.fit(X,y)
print('Intercept cua mo hình bac 2')
print(model.intercept_)
print('Coeffient cua mo hình bac 2')
print(model.coef_)
print('Do chinh xac cua mo hình bac 2')
print(model.score(X,y))
```

```python
range = np.arange(1.5,5.5,0.1)
predictor = pd.DataFrame([range, range**2]).T
X = predictor
y_pred = model.predict(X)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9),
          color = 'black', xlim = (0,7))
plt.plot(range, y_pred, color = 'blue')
```

```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

data = pd.read_csv('50_Startups.csv')
data.head()
sns.pairplot(data)

X = data.drop(['Profit','State'], axis = True)
y = data['Profit']
X.head
```

```python
corr = data.corr() 
sns.heatmap(corr, annot = True)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test).round(1)

calculation = pd.DataFrame(np.c_[y_test, y_pred],
                           columns = ['Original Profits', 'Predicted Profits'])
calculation.head()
```

```python
print(regressor.score(X_train, y_train))
```

```python
print(regressor.score(X_test, y_test))
```
